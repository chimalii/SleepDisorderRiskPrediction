## **1. Carga de modelo y arfactos**

```python
import os
# Ocultar advertencias de TensorFlow en la terminal
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import pandas as pd
import numpy as np
import tensorflow as tf
import joblib

model = tf.keras.models.load_model('classifier_model.keras')
scaler = joblib.load('scaler.joblib')
expected_columns = joblib.load('columnas_esperadas.joblib')

# Clases objetivo
CLASS_NAMES = ['Healthy', 'Mild', 'Moderate', 'Severe']
input_data = {}

print(">> Modelo y artefactos cargados...")
```

## **2. Ingresar variables numéricas**

```python
print("\n" + "="*50)
print(" PREDICTOR DE RIESGO DE TRASTORNO DEL SUEÑO (MLP)")
print("="*50 + "\n")

num_prompts = {
    'age': '18-69 años',
    'bmi': '16-45',
    'sleep_duration_hrs': '3-10.5 horas',
    'sleep_quality_score': '1 = muy mala / 10 = excelente',
    'rem_percentage': '10-30%',
    'deep_sleep_percentage': '5-30%',
    'sleep_latency_mins': '1-58 minutos',
    'wake_episodes_per_night': '0-8 veces',
    'alcohol_units_before_bed': '0-6 unidades',
    'screen_time_before_bed_mins': '2-180 minutos',
    'steps_that_day': '500-20,000 pasos',
    'nap_duration_mins': '0-116 minutos',
    'stress_score': '1 = muy bajo / 10 = extremo',
    'work_hours_that_day': '0-18 horas',
    'heart_rate_resting_bpm': '45-99 latidos por minuto',
    'room_temperature_celsius': '15-28°C',
    'weekend_sleep_diff_hrs': '-1 a 3 horas',
    'caffeine_mg_before_bed': '0, 40, 80, 100, 150, 200, 300, 400 mg'
}

for col, desc in num_prompts.items():
        while True:
            try:
                val = input(f"Ingresa {col} ({desc}): ")
                input_data[col] = [float(val)]
                break
            except ValueError:
                print("Valor numérico inválido.")
```

## **3. Ingresar variables categóricas nominales**

```python
print("\n--- Variables Categóricas ---")

# Diccionario con las opciones exactas que existen en tu dataset
cat_options = {
    'gender': ["Female", "Male", "Other"],
    'occupation': ["Student", "Software Engineer", "Nurse", "Manager", "Teacher", "Doctor",
                    "Retired", "Sales", "Freelancer", "Driver", "Homemaker", "Lawyer"],
    'country': ["USA", "India", "Japan", "UK", "Brazil", "Germany", "Canada", "France",
                "Australia", "Italy", "South Korea", "Mexico", "Netherlands", "Spain", "Sweden"],
    'chronotype': ["Neutral", "Morning", "Evening"],
    'mental_health_condition': ["Healthy", "Anxiety", "Depression", "Both"],
    'season': ["Spring", "Summer", "Autumn", "Winter"]
}

for col, options in cat_options.items():
    print(f"\nOpciones para {col}:")
    # Imprimir el submenú con números empezando desde el 1
    for i, opt in enumerate(options, 1):
        print(f"  {i}. {opt}")

    while True:
        try:
            sel = int(input(f">> Selecciona el número para {col}: "))
            # Validar que el número esté dentro del rango del menú
            if 1 <= sel <= len(options):
                # Asignar el texto correspondiente al número elegido (restando 1 por el índice)
                input_data[col] = [options[sel - 1]]
                break
            else:
                print(f"Opción fuera de rango. Ingresa un número entre 1 y {len(options)}.")
        except ValueError:
            print("Entrada inválida. Debes ingresar un número.")
```

## **4. Ingresar variables categóricas binarias**

```python
print("\n--- Variables Binarias ---")
bin_prompts = {
    'exercise_day': '1 = sí / 0 = no',
    'sleep_aid_used': '1 = sí / 0 = no',
    'shift_work': '1 = sí / 0 = no',
    'day_type': '1 = Weekday / 0 = Weekend'
}

for col, desc in bin_prompts.items():
    while True:
        try:
            val = input(f"Ingresa {col} ({desc}): ")
            input_data[col] = [int(val)]
            if input_data[col][0] not in [0, 1]:
                raise ValueError
            break
        except ValueError:
            print("Ingresa únicamente 1 o 0.")
```

## **5. Realizar predicción**

```python
print("\nProcesando datos y aplicando transformaciones del ETL...")
df_input = pd.DataFrame(input_data)

# Definir listas de columnas para el preprocesamiento
num_cols = list(num_prompts.keys())
cat_cols = list(cat_options.keys())

# 1. Estandarización de variables numéricas con el scaler entrenado
df_input[num_cols] = scaler.transform(df_input[num_cols])

# 2. One-Hot Encoding respetando drop_first=True
df_input = pd.get_dummies(df_input, columns=cat_cols, drop_first=True, dtype=int)

# 3. Alinear la estructura con las 57 características esperadas
df_input = df_input.reindex(columns=expected_columns, fill_value=0)

# Convertir a arreglo de numpy para el modelo de Keras
X_pred = df_input.to_numpy(dtype=np.float32)

# Predicción
prediction_probs = model.predict(X_pred, verbose=0)[0]
predicted_class_idx = np.argmax(prediction_probs)
predicted_class = CLASS_NAMES[predicted_class_idx]

print("\n" + "="*50)
print(" RESULTADO DE LA PREDICCIÓN")
print("="*50)
print(f"➜ Riesgo Predicho: ** {predicted_class.upper()} **\n")

print("Desglose de probabilidades:")
for i, cls in enumerate(CLASS_NAMES):
  print(f"  - {cls}: {prediction_probs[i]*100:.2f}%")
```
